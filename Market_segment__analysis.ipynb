# EDA (Exploratory Data Analysis)

## Distribution of age

```python
age = sns.distplot(df['Age'], hist_kws = {'rwidth':0.8, 'color': 'red'}, bins = np.array([ 0, 20, 40, 60, 100]))
plt.title('Distribution of Age')
for bars in age.containers:
    age.bar_label(bars)
```

Conclusion 1: Mostly people shopping from this mall belongs to age group 40-60 followed by 20-40

## Gender Distribution

```python
px.pie(df, names = 'Gender', title = 'Gender Distribution')
```

Comclusion 2: Maximum purchases are made by 'Males\'

## Most commonly purchased items



```python
item_counts = df['Item Purchased'].value_counts().reset_index()
```

```python
item_counts.columns = ['Item Purchased', 'Count']
```

```python
px.bar(item_counts, x = 'Item Purchased', y = 'Count', title = 'Most Commonly Purchased Items')
```

Conclusion 3: 'blouse' is the most purchased item followed by 'Jwellery' and 'Pants' while the jeans are least preferred one

## Seasoning shopping distribution

```python
px.pie(df, names = 'Season', title = 'Seasoning Shopping Distribution')
```

Conclusion 4: People mostly shop in the season of 'Spring' and 'Fall'

## Location distribution

```python
item_counts = df['Location'].value_counts().reset_index()
```

```python
px.bar(item_counts, x = 'Location', y = 'count', title = 'Location Distribution', color = 'Location')
```

Conclusion 5: Mostly the people prefer shopping from 'Montana' outlet followed by 'California' outlet

## Preferred payment method

```python
px.pie(df, names = 'Payment Method', title = 'Preferred Payment Method')
```

Conclusion: "Credit Card' and 'Venmo' are the most preferred methods

## Frequency of Purchase Distribution

```python
px.histogram(df, x = 'Frequency of Purchases', title = 'Frequency of Purchase Distribution')
```

Conclusion 7: Most people prefer shopping after 3 monnths

## Promo code usage distribution

```python
px.pie(df, names = 'Promo Code Used', title = 'Promo Code Usage Distribution')
```

Conclusion 8: Majority people do shopping without using any Promo Code


## Shipping Type Distribution

```python
px.pie(df, names = 'Shipping Type', title = 'Shipping Type Distribution')
```

Conclusion 9: Mostly people do 'Free Shipping'




## Average review rating genderwise

```python
 px.box(df, x = 'Gender', y = 'Review Rating', title = 'Average Review Rating Genderwise')
```

Conclusion 10: Average rating by Male > average rating by female

## Color preference distribution

```python
px.pie(df, names = 'Color', title = 'Color Preference Distribution')
```

Conclusion 11: Mostly people shop for 'olive' color followed by 'yellow'

## Category by gender and payment method

```python
px.sunburst(df, path = ['Gender', 'Payment Method', 'Category'], title = 'Category by Gender and Payment Method')
```

Conclusion 12:


1.   Mostly the shopping is done by 'Male'
2.   Mostly the men/female shop for Clothing followed
